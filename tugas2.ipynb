```python
# Nama : Rifa'i Nur Ahsani
# Kelas: M2025E
# NIM  : 25030214134

z = int(input("Masukkan angka : "))
q = 0

if z % 2 == 0:
    y = "Genap"
else:
    y = "Ganjil"

for i in range(1,z+1):
    if z % i == 0:
        q += 1

if q == 2:
    w = "Prima"
else:
    w = "Non Prima"

print(f"{z} adalah Bilangan {y} dan {w}")
```
